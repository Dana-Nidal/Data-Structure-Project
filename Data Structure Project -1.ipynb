```python
# Import necessary libraries
from collections import Counter, deque

# Stack implementation
class TextStack:
    def __init__(self):
        ##Initialize an empty list to represent the stack
        self.items = []
    def add(self, item):
        ## Pushes an item onto the stack.
        self.items.append(item)
    def remove(self):
        ## Removes and returns the top item from the stack.
        return self.items.pop() if not self.is_empty() else None
    
    def is_empty(self):
        ##  Checks if the stack is empty.
        return len(self.items) == 0

# Queue implementation
class TextQueue:
    def __init__(self):
        # Initialize an empty deque to represent the queue
        self.items = deque()

    def add(self, item):
        # Enqueue an item at the end of the queue
        self.items.append(item)

    def remove(self):
        # Dequeue an item from the front of the queue, if not empty
        return self.items.popleft() if not self.is_empty() else None

    def is_empty(self):
        # Check if the queue is empty
        return len(self.items) == 0

# Linked list node 
class TextNode:
    def __init__(self, data):
    # Initialize a node with data and a pointer to the next node
        self.data = data
        self.next = None

# Linked list implementation
class TextLinkedList:
    def __init__(self):
        # Initialize an empty linked list with no head node
        self.head = None

    def add(self, data):
        # Add a new node with the given data to the end of the linked list
        new_node = TextNode(data)
        if not self.head:
            self.head = new_node # Set the new node as the head if the list is empty
        else:
            current = self.head
            while current.next:
                current = current.next # Traverse to the end of the list
            current.next = new_node

    def traverse(self):
        # Traverse the linked list and return all node data in a list
        result = []
        current = self.head
        while current:
            result.append(current.data)
            current = current.next
        return result

# Main text processor Class 
class TextProcessor:
    def __init__(self):
        # Initialize an empty string to hold user input
        self.text = ""

    def get_user_input(self):
         # Prompt the user to enter a string of text
        self.text = input("Enter a string of text: ")

    def analyze_characters(self):
        # Count and display the frequency of each character in the input text
        char_array = list(self.text)
        char_frequency = Counter(char_array)
        print("Character Frequencies:")
        for char, freq in char_frequency.items():
            print(f"{char}: {freq}")

    def analyze_words(self):
        # Count and display the frequency of each word in the input text
        words = self.text.split()
        unique_words = set(words)
        word_frequency = Counter(words)
        print("\nWord Frequencies:")
        for word, freq in word_frequency.items():
            print(f"{word}: {freq}")
        return unique_words

    def reverse_text(self):
        # Reverse the input text using a stack
        stack = TextStack()
        for char in self.text:
            stack.add(char) # Push each character onto the stack
        reversed_text = ""
        while not stack.is_empty():
            reversed_text += stack.remove()# Pop characters to form the reversed text
        print("\nReversed String:")
        print(reversed_text)

    def display_fifo_words(self):
        # Display words in First-In, First-Out (FIFO) order using a queue
        queue = TextQueue()
        words = self.text.split() # Split the input into words
        for word in words:
            queue.add(word) # Enqueue each word
        print("\nFIFO Ordering of Words:")
        while not queue.is_empty():
            print(queue.remove(), end=" ") # Dequeue words and print them
        print()

    def linked_list_traversal(self):
        # Add words to a linked list and display them in order
        words = self.text.split() # Split the input into words
        linked_list = TextLinkedList()
        for word in words:
            linked_list.add(word) # Add each word as a node in the linked list
        print("\nLinked List Traversal:")
        print(" -> ".join(linked_list.traverse())) # Traverse and print the linked list

# Call all the Features 
if __name__ == "__main__":
     # Initialize the text processor
    processor = TextProcessor()
    # Run text processing functionalities step by step
    processor.get_user_input() # Get input from the user
    processor.analyze_characters() # Analyze and display character frequencies
    processor.analyze_words()# Analyze and display word frequencies
    processor.reverse_text() # Reverse the input text using a stack
    processor.display_fifo_words() # Display words in FIFO order using a queue
    processor.linked_list_traversal() # Traverse and display words using a linked list
```
